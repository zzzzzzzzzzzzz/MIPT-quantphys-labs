## Опыт Франка-Герца
Методом электронного возбуждения измеряется энергия 1-ого уровня атома гелия в динамическом и статическом режимах

```python
# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import scipy.interpolate as scpint
```

Подгрузим данные в статическом режиме, 1.csv, 2.csv, 3.csv соответствуют 4,6 и 8 В соответственно (значение задерживающего напряжения)

```python
four_volts = pd.read_csv("1.csv")
six_volts = pd.read_csv("2.csv")
eight_volts = pd.read_csv("3.csv")
```

Переведём ток из делений в мА и добавим столбцы систематических ошибок

```python
four_volts = four_volts.join(pd.DataFrame({'$I_{к}$, мА':[2.5 * four_volts.values.T[1][i] for i in range(four_volts.shape[0])], 'xerr':[0.005 for i in range(four_volts.shape[0])], 'yerr': [2.5 for i in range(four_volts.shape[0])]}))
four_volts
```

```python
six_volts = six_volts.join(pd.DataFrame({'$I_{к}$, мА':[2.5 * six_volts.values.T[1][i] for i in range(six_volts.shape[0])], 'xerr':[0.005 for i in range(six_volts.shape[0])], 'yerr': [2.5 for i in range(six_volts.shape[0])]}))
six_volts
```

```python
eight_volts = eight_volts.join(pd.DataFrame({'$I_{к}$, мА':[2.5 * eight_volts.values.T[1][i] for i in range(eight_volts.shape[0])], 'xerr':[0.005 for i in range(eight_volts.shape[0])], 'yerr': [2.5 for i in range(eight_volts.shape[0])]}))
eight_volts
```

Построим графики зависимости $I_{к} = I_{к}(V_1)$ при разных задерживающих напряжениях

```python
plt.figure(num=None, figsize=(12, 6), dpi=80, facecolor='w', edgecolor='k')
plt.subplot()

yFourVolts = four_volts.values.T[2]
xFourVolts = four_volts.values.T[0]
xerrFourVolts = four_volts.values.T[3]
yerrFourVolts = four_volts.values.T[4]

ySixVolts = six_volts.values.T[2]
xSixVolts = six_volts.values.T[0]
xerrSixVolts = six_volts.values.T[3]
yerrSixVolts = six_volts.values.T[4]

yEightVolts = eight_volts.values.T[2]
xEightVolts = eight_volts.values.T[0]
xerrEightVolts = eight_volts.values.T[3]
yerrEightVolts = eight_volts.values.T[4]

#sugar
plt.errorbar(xFourVolts, yFourVolts, xerr=xerrFourVolts, yerr=yerrFourVolts, fmt='k.', color='r')
plt.errorbar(xSixVolts, ySixVolts, xerr=xerrSixVolts, yerr=yerrSixVolts, fmt='k.', color='g')
plt.errorbar(xEightVolts, yEightVolts, xerr=xerrEightVolts, yerr=yerrEightVolts, fmt='k.', color='b')
plt.xlabel(u'$V_{а}$, B')
plt.ylabel(u'$I_{к}$, mA')
plt.title(u'The dependence of electricity on voltage of anode')
plt.grid(True)

#plot
x_smooth = np.linspace(xFourVolts.min(), xFourVolts.max(), 300)
y_smooth = scpint.spline(xFourVolts, yFourVolts, x_smooth)
plt.plot(x_smooth, y_smooth, color='r')
x_smooth = np.linspace(xSixVolts.min(), xSixVolts.max(), 300)
y_smooth = scpint.spline(xSixVolts, ySixVolts, x_smooth)
plt.plot(x_smooth, y_smooth, color='g')
x_smooth = np.linspace(xEightVolts.min(), xEightVolts.max(), 300)
y_smooth = scpint.spline(xEightVolts, yEightVolts, x_smooth)
plt.plot(x_smooth, y_smooth, color='b')
plt.legend(['4 Volts',
            '6 Volts',
            '8 Volts'],    # список легенды
            loc='upper left')    # положение легенды
plt.show()
```

Табличное значение энергии возбуждения первого уровня атома гелия 19.82 эВ

```python
plt.figure(num=None, figsize=(12, 6), dpi=80, facecolor='w', edgecolor='k')
plt.subplot()

plt.errorbar(xFourVolts, yFourVolts, xerr=xerrFourVolts, yerr=yerrFourVolts, fmt='k.', color='r')
plt.xlabel(u'$V_{а}$, B')
plt.ylabel(u'$I_{к}$, mA')
plt.title(u'The dependence of electricity on voltage of anode')
plt.grid(True)

plt.plot(four_volts.values[7][0], four_volts.values[7][2], 'go')
plt.plot(four_volts.values[8][0], four_volts.values[8][2], 'go')
lefterr = (four_volts.values[8][0] - four_volts.values[7][0])/2 
left =  lefterr + four_volts.values[7][0]

plt.plot(four_volts.values[17][0], four_volts.values[17][2], 'go')
plt.plot(four_volts.values[18][0], four_volts.values[18][2], 'go')
righterr = (four_volts.values[18][0] - four_volts.values[17][0])/2 
right =  righterr + four_volts.values[17][0]


plt.show()

print "Result"
four_res = round(right - left,2)
four_err = round(lefterr+righterr, 2)
print str(four_res) + " ± " + str(four_err) + "эВ"
```

```python
plt.figure(num=None, figsize=(12, 6), dpi=80, facecolor='w', edgecolor='k')
plt.subplot()

plt.errorbar(xEightVolts, yEightVolts, xerr=xerrEightVolts, yerr=yerrEightVolts, fmt='k.', color='b')
plt.xlabel(u'$V_{а}$, B')
plt.ylabel(u'$I_{к}$, mA')
plt.title(u'The dependence of electricity on voltage of anode')
plt.grid(True)

plt.plot(eight_volts.values[4][0], eight_volts.values[4][2], 'yo')
plt.plot(eight_volts.values[6][0], eight_volts.values[6][2], 'yo')
lefterr = (eight_volts.values[6][0] - eight_volts.values[4][0])/2 
left =  lefterr + eight_volts.values[4][0]

plt.plot(eight_volts.values[15][0], eight_volts.values[15][2], 'yo')
plt.plot(eight_volts.values[16][0], eight_volts.values[16][2], 'yo')
righterr = (eight_volts.values[16][0] - eight_volts.values[15][0])/2 
right =  righterr + eight_volts.values[15][0]


plt.show()

print "Result"
eight_res = round(right - left, 2)
eight_err = round(lefterr + righterr, 2)
print str(eight_res) + " ± " + str(eight_err) + "эВ"
```
